```python
import pandas as pd

df_pd = pd.read_csv("Data_Science_Fields_Salary_Categorization.csv")
df_pd.head()
```

```python
df_pd['Salary_In_Rupees'] = df_pd['Salary_In_Rupees'].str.replace(',', '').astype(float)
```

```python
df_pd.head()
```

```python
df_pd['Salary_In_Dollars'] = df_pd['Salary_In_Rupees'] / 90.60
```

```python
df_pd.head()
```

```python
df_pd['Experience'] = sqldf("""
                            SELECT
                              CASE Experience
                                WHEN 'EN' THEN 'Junior'
                                WHEN 'MI' THEN 'Pleno'
                                WHEN 'SE' THEN 'Senior'
                                WHEN 'EX' THEN 'Executive'
                                END AS Experience
                            FROM
                              df_pd
                            """)
```

```python
df_pd.head()
```

```python
df_pd = sqldf(
              """
              SELECT
                working_year,
                Designation,
                Experience,
                Employment_Status,
                Salary_In_Dollars,
                Employee_Location,
                Company_Location,
                Company_Size,
                Remote_Working_Ratio
              FROM
                df_pd
              """
              )
```
